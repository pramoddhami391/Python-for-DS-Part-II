create a  table  for Books with fields:title,author,year.
Insert 3 books and query by author.
Update a book's year.
Delete a book.

```python
import sqlite3
conn=sqlite3.connect("Library.db")
cursor=conn.cursor()

```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Books(
               title TEXT PRIMARY KEY,
               author TEXT,
               year INTEGER
               
               )
               """)
```

```python
cursor.execute("INSERT INTO Books (title,author,year) VALUES(?,?,?)",("Mastery","Robert Greene",2012))
conn.commit()
```

```python
cursor.execute("INSERT INTO Books (title,author,year) VALUES(?,?,?)",("MasteryData communications and networking","Forouzan, B. A.",2012))
conn.commit()
```

```python
cursor.execute("INSERT INTO Books (title,author,year) VALUES(?,?,?)",("Python for Data Analysis","McKinney, W",2022))
conn.commit()
```

```python
cursor.execute("UPDATE Books SET title='Data communications and networking' WHERE title='MasteryData communications and networking' ")
conn.commit()
```

```python
cursor.execute("SELECT * FROM Books where author='Robert Greene'")
rows=cursor.fetchall()
for row in rows:
    print(rows)
```

```python
cursor.execute("UPDATE Books SET year='2010' WHERE title='Data communications and networking'")
conn.commit()
```

```python
cursor.execute("DELETE FROM Books WHERE title='Data communications and networking'")
conn.commit()
```
