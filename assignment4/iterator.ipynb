1.Write an iterator that returns the first n even numbers.

```python
class Even:
    def __init__(self,limit):
        self.limit=limit
        self.current=0
        self.count=0

    def __iter__(self):
        return self
    
    def __next__(self):
       if self.count>=self.limit:
           raise StopIteration
       
       value=self.current
       self.current+=2
       self.count+=1
       return value
        
n_even_numbers=Even(5)
for num in n_even_numbers:
    print(num)

# using generator

# def even(n):
#     for i in range(n*2):
#         if i%2==0:
#             yield i
# for num in even(10):
#     print(num)
```

2.Write a generator that yields the Fibonacci sequence up to n terms.

```python
def squares(n):
    for i in range(n):
        yield i*i

for square in squares(5):
    print(square)
```

```python
def fibo(n):
    a,b=0,1
    for i in range(n):
        
        yield a
        a,b=b,b+a
        

for seq in fibo(7):
    print(seq)
```

4.Write a generator to read a CSV file line by line

```python
def read_large_file(filepath="SUV_purchase.csv"):
    with open(filepath) as f:
        for line in f:
            yield line.strip()

for line in read_large_file('SUV_purchase.csv'):
    print(line)
```

5.Implement an iterator class that iterates over a range in reverse order.

```python
class Reverse:
    def __init__(self,initial,final):
        self.initial=initial
        self.final=final
        
        

    def __iter__(self):
        return self
    
    def __next__(self):
       if self.initial<=self.final:
           raise StopIteration
       
       value=self.initial
       self.initial-=1
       return value
          
range_rev=Reverse(6,1)
for num in range_rev:
    print(num)
```

```python
import tracemalloc

def decorator(func):
    def wrapper(*args, **kwargs):
        tracemalloc.start()

        result = func(*args, **kwargs)

        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        print(f"Current memory usage: {current / 1024:.2f} KB")
        print(f"Peak memory usage: {peak / 1024:.2f} KB")

        return result
    return wrapper


@decorator
def func():
    data = [i for i in range(10)]
    return data


func()
```

8.Write a generator that yields prime numbers indefinitely.

```python
import math
def is_prime(n):
    if n<2:
        return False
    
    for i in range(2,int(math.sqrt(n))+1):
        if n%i==0:
            return False
    return True

def prime_generator():
    num=2
    while True:
        if is_prime(num):
            yield num
        num+=1

prime_gen=prime_generator()
print("First 10 prime numbers are:")
for _ in range(10):
    print(next(prime_gen),end=" ")
```

Create a decorator to cache results of a function (memoization).

```python
import functools
def memorize(func):
    cache={}
    @functools.wraps(func)
    def wrapper(*args):
        if args in cache:
            print(f"Fetching result for {args} from cache...")
            return cache[args]
        
        result=func(*args)
        cache[args]=result
        return result
    return wrapper
```

```python
import time

@memorize
def slow_fib(n):
    if n<2:
        return n
    return slow_fib(n-1)+slow_fib(n-2)

# testing
start=time.time()
print(f"Result: {slow_fib(35)}")
end=time.time()

print(f"Time taken: {end-start:.5f} seconds")

# calling again to see the cache in action

print("Calling again")
start=time.time()
print(f"Result: {slow_fib(35)}")
end=time.time()
print(f"Time taken: {end- start:.5f} seconds")
```

12.Implement a decorator to log execution time and input arguments.

```python
import time
import functools

def log_execution(func):
    @functools.wraps(func)
    def wrapper(*args,**kwargs):
        args_repr=[repr(a) for a in args]
        kwargs_repr=[f"{k} = {v!r} " for k,v in kwargs.items()]

        signature=", ".join(args_repr+ kwargs_repr)

        print(f"--- Calling {func.__name__} ({signature}) ---")

        start=time.perf_counter()
        result=func(*args, **kwargs)

        end=time.perf_counter()

        duration=end-start

        print(f"---{func.__name__} finished in {duration:.4f} seconds ---")
        return result
    
    return wrapper

@log_execution
def process_data(name, delay=1):
    time.sleep(delay)
    return f"Data for {name} processed."

process_data("sensor_A",delay=0.4)
```

13.Write an iterator that yields only unique items from a list.

```python

class Unique:
    def __init__(self,lst):
        self.lst=lst
        self.index=0
        self.seen=set()

       

    def __iter__(self):
        return self
    
    def __next__(self):
        while self.index<len(self.lst):
            value=self.lst[self.index]
            self.index+=1

            if value not in self.seen:
                self.seen.add(value)
                return value
            
        raise StopIteration
       
lst=[1,2,3,2]
u=Unique(lst)
for unique_value in u:
    print(unique_value)
```

14.Implement a decorator to check if inputs of a function are positive numbers.

```python
def positive_only(func):
    def wrapper(*args, **kwargs):
        for value in args:
            if not isinstance(value, (int, float)) or value <= 0:
                raise ValueError("All inputs must be positive numbers")

        for value in kwargs.values():
            if not isinstance(value, (int, float)) or value <= 0:
                raise ValueError("All inputs must be positive numbers")

        return func(*args, **kwargs)
    return wrapper


@positive_only
def add(a, b):
    return a + b

print(add(3, 4))     
print(add(0, 5))   #raise value error
```
